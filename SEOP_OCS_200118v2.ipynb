```python
###### WSCH LOGIN ASSISTANT VER 20200118

import tkinter as tk
import os
import win32com.client
import time
import win32gui, win32con
import pyautogui as pg

##### Block Ad pop-up
ad = False
def adblock(hwnd, lParam):
    if 'http://' in win32gui.GetWindowText(hwnd):
        win32gui.PostMessage(hwnd, win32con.WM_CLOSE, 0, 0)
        ad = True

##### Close all IE, then open IE 
def close_open_IE():
    os.system('taskkill /f /im iexplore.exe')
    global ie
    ie = win32com.client.Dispatch('InternetExplorer.Application')

##### Waiting until ready in IE
def waiting_IE():
    while ie.ReadyState !=4:
        time.sleep(0.5)
     
##### WSCH Login and Activate adblock    
def WSCH_login(ID, PW):
    close_open_IE()
    ie.Visible = 1
    ie.Left = -1920
    ie.Top = 0
    ie.TheaterMode = 1
    ie.Navigate ('http://128.9.1.11')
    
    waiting_IE()
    
    ie.document.getElementById('txtEmpNo').value = ID
    ie.document.getElementById('txtPswd').value = PW
    ie.document.parentWindow.execScript ('login();')
    
    for t in range(10):
        win32gui.EnumWindows(adblock, None)
        time.sleep(0.2)
        if ad == True:
            break
    waiting_IE()

##### Execute login
def assi_login():
    wsch_id = str(et_id.get())
    wsch_pw = str(et_pw.get())
    WSCH_login(wsch_id, wsch_pw)

    if int(con_option.get()) == 1:
        ie.document.parentWindow.execScript ('window.frames[0].document.getElementById("a_17").click()')
        waiting_IE()
        ie.document.parentWindow.execScript ('window.frames[1].retrieve_e_prcp();')
        
    elif int(con_option.get()) == 2:
        ie.document.parentWindow.execScript ('window.frames[0].document.getElementById("a_14").click()')

    elif int(con_option.get()) == 3:
        ie.document.parentWindow.execScript ('window.frames[0].document.getElementById("a_13").click()')
    
##### See menu of the cafeteria
def cafeteria():
    close_open_IE()
    ie.Navigate ("http://gw.ywmc.or.kr/jsp/bbs/board/bbslist.jsp?K=00PtdF3r&BRDID=0000009fn&OCSUSER=Y")
    ie.Visible = 1
    ie.Left = -1920
    ie.Top = 0
    ie.Width = 1920
    ie.Height = 1080

##### EMR sign in the ED
def assi_emr_sign():
    emr_note = str(emr_sign.get())
    pg.click((-990, 896))

    time.sleep (2)

    pg.press(["tab", "tab", "tab", "tab", "tab"])
    pg.typewrite(emr_note)
    pg.press(["tab", "down"]) # GW

    if int(emr_option.get()) == 1: # ICU
        pg.press("down")
    elif int(emr_option.get()) == 2: # DC
        pg.press("up")
        
    pg.keyDown("shift")
    pg.press(["tab", "tab", "tab", "tab", "tab", "tab", "tab", "tab", "tab", "space"])
    pg.keyUp("shift")
    time.sleep(0.5)
    pg.hotkey("alt", "f4")
    time.sleep(0.5)

    pg.click((-1786, 162))
    time.sleep(0.1)
    pg.click((-1882, 249))
    time.sleep(0.5)
    pg.doubleClick((-1566, 243))
    time.sleep(0.5)
    pg.doubleClick((-1099, 556))
    pg.click((-726, 928))    

    
##### TKINTER CONFIGURATIONS
ww = tk.Tk()
ww.geometry("300x240-1920+70")
ww.resizable(0, 0)
ww.title("WSCH LOGIN - SEOP")
ww.attributes("-topmost", True)

## [Label] ID, PW, caution
lb_id = tk.Label(ww, text = "ID", font = ("Times New Roman", "15")).place(x = 10, y = 10)
lb_pw = tk.Label(ww, text = "PW", font = ("Times New Roman", "15")).place(x = 10, y = 40)
lb_caution = tk.Label(ww, text =r"※ 열려있는 IE가 전부 닫힙니다!").place(x=10, y=120)

## [Entry, Button] Login
other_id = tk.StringVar()
other_pw = tk.StringVar()
other_id.set("2191039")
other_pw.set("789789789+")

et_id = tk.Entry(ww, textvariable = other_id, width = 15, font = ("Times New Roman", "15"))
et_id.place(x = 50, y = 10)
et_pw = tk.Entry(ww, textvariable = other_pw, width = 15, font = ("Times New Roman", "15"), show="*")
et_pw.place(x = 50, y = 40)

bt_login = tk.Button(ww, text="Login", bg = "green", bd = 5, command = assi_login, font = ("Times New Roman", "15")).place(x = 50, y = 70)
bt_menu = tk.Button(ww, text="Menu", bg = "red", bd = 5, command = cafeteria, font = ("Times New Roman", "15")).place(x = 120, y = 70)

## [RB] Option for OCS connection: Main, ED, Adm, OPD
con_option = tk.IntVar()
rb_main = tk.Radiobutton(ww, text="Main", variable = con_option, value = "0", font = ("Times New Roman", "15"))
rb_ed = tk.Radiobutton(ww, text="ED", variable = con_option, value = "1", font = ("Times New Roman", "15"))
rb_adm = tk.Radiobutton(ww, text="Adm", variable = con_option, value = "2", font = ("Times New Roman", "15"))
rb_cons = tk.Radiobutton(ww, text="OPD", variable = con_option, value = "3", font = ("Times New Roman", "15"))

rb_main.place (x = 220, y = 0)
rb_ed.place (x = 220, y = 25)
rb_adm.place (x = 220, y = 50)
rb_cons.place (x= 220, y = 75)
rb_main.select()

## [ENTRY, Btn, RB] Option for EMR sign in the ED: GW, ICU, DC
emr_default = tk.StringVar()
emr_default.set("lab, imaging study")

emr_sign = tk.Entry(ww, textvariable = emr_default)
emr_sign.place(x = 10, y = 150, width = 270)

bt_emr = tk.Button(ww, text="EMR sign", bg = "white", bd = 5, command = assi_emr_sign, font = ("Times New Roman", "15")).place(x = 10, y = 170)

emr_option = tk.IntVar()
rb_emr_GW = tk.Radiobutton(ww, text="GW", variable = emr_option, value = "0", font = ("Times New Roman", "10"))
rb_emr_ICU = tk.Radiobutton(ww, text="ICU", variable = emr_option, value = "1", font = ("Times New Roman", "10"))
rb_emr_DC = tk.Radiobutton(ww, text="DC", variable = emr_option, value = "2", font = ("Times New Roman", "10"))
rb_emr_GW.place (x = 220, y = 170)
rb_emr_ICU.place (x = 220, y = 190)
rb_emr_DC.place (x = 220, y = 210)
rb_emr_GW.select()

ww.mainloop()
```
